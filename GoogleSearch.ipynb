```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.io.json import json_normalize

sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
pytrend=TrendReq()
```

```python
#Keywords provincias

user_kw_list =['Python', 'Machine Learning', 'Google Cloud', "Deep Learning", "SQL"]
user_cat = 0 
user_timeframe = 'today 12-m' 
user_geo = 'BR'
```

```python
pytrend.build_payload(kw_list=user_kw_list, cat=user_cat, timeframe=user_timeframe, geo=user_geo)
```

```python
interest_over_time_df = pytrend.interest_over_time()
print(interest_over_time_df.head())
interest_over_time_df.to_csv('Search_BR.csv')
```

```python
interest_over_time_df.plot()
plt.title('Search Time Series Brasil', fontsize = 20)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Search', fontsize = 15)
plt.show()
plt.savefig('search_br.png')

```
